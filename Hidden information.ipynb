```python
#! /usr/bin/env python
# -*- coding: utf-8 -*-
%matplotlib inline
%config InlineBackend.figure_format = 'svg'

import numpy as np
import pandas as pd
import seaborn as sns; sns.set(style="white")
import matplotlib.pyplot as plt

np.random.seed(seed=42)

# Generating data (sets)
# Data set A
s = np.random.normal(20.0, 1.5, 50)
t = np.random.normal(15.0, 1.0, 70)
A = np.append(s, t)

# Data set B
s = np.random.normal(25.0, 3.0, 5)
t = np.random.normal(18.0, 1.0, 10)
B = np.append(s, t)

# Data set C
s = np.random.normal(17.0, 1.0, 150)
t = np.random.normal(9.0, 2.5, 300)
C = np.append(s, t)

# Data set D
D = np.random.normal(11.0, 1.0, 200)


# Store data in df
A, B, C, D = pd.Series(A), pd.Series(B), pd.Series(C), pd.Series(D)
df = pd.concat([A,B,C,D], ignore_index=True, axis=1)
df.columns=['A', 'B', 'C', 'D']

df.head(5)
```

```python
colors = ["amber", "greyish", "faded green", "dusty purple","windows blue"]
sns.set_palette(sns.xkcd_palette(colors))
# Reference for xkcd colors: https://xkcd.com/color/rgb/
```

```python
def figure_properties():
    plt.figure(figsize=[5,5])
    plt.ylim(0,30)
    plt.xlabel('Data set')
    plt.ylabel('Arbitrary units')
    plt.rcParams['axes.facecolor'] = [0.98,0.98,0.98] # hellgrau
```

```python
figure_properties()

sns.barplot(data=df, ci='sd', capsize=.2, errwidth=1.5, edgecolor='.25', linewidth=1.5);
```

```python
figure_properties()
sns.boxplot(data=df);
```

```python
figure_properties()

sns.boxplot(data=df)
sns.swarmplot(data=df, size=1.5, color='0');
```

```python
factor = 5
x_min = 0
x_max = 20

f, axes = plt.subplots(1, 4, figsize=(5, 5), sharey=True)
ax1 = sns.distplot(df.A.dropna(), bins=int((df.A.max()-df.A.min()))*factor, kde=False, color='black', ax=axes[0], vertical=True, hist_kws=dict(alpha=1))
ax2 = sns.distplot(df.B.dropna(), bins=int((df.B.max()-df.B.min()))*factor, kde=False, color='black', ax=axes[1], vertical=True, hist_kws=dict(alpha=1))
ax3 = sns.distplot(df.C.dropna(), bins=int((df.C.max()-df.C.min()))*factor, kde=False, color='black', ax=axes[2], vertical=True, hist_kws=dict(alpha=1))
ax4 = sns.distplot(df.D.dropna(), bins=int((df.D.max()-df.D.min()))*factor, kde=False, color='black', ax=axes[3], vertical=True, hist_kws=dict(alpha=1))

ax1.set_xlim(x_min, x_max)
ax2.set_xlim(x_min, x_max)
ax3.set_xlim(x_min, x_max)
ax4.set_xlim(x_min, x_max)

plt.ylim(0,30)

plt.tight_layout();
```

```python
figure_properties()

sns.violinplot(data=df, inner='box', cut=2);
```
